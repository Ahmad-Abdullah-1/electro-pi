READING THE IMPORTED DATA 

```python
 
def DataReading(file_name, file_ext):
    if file_ext == 'csv':
        df = pd.read_csv(file_name, encoding='latin-1')
    elif file_ext == 'excel':
        df = pd.read_excel(file_name, encoding='latin-1')
    elif file_ext == 'json':
        sf = pd.read_json(file_name ,encoding='latin-1')
    else:
        raise ValueError("file not acceptable")

    return df
```

```python
df = DataReading(file_name, file_ext)
```

HANDLING MISSING VALUES

```python
def HandlingMissingValues(df, command ='remove'):
    if command == 'remove':
        df = df.dropna()
    elif command == 'impute':  # Use 'elif' instead of 'else' for multiple conditions
        df = df.fillna(df.mean())  # Fill null values with mean

    return df
```

```python
df = HandlingMissingValues(df, command = 'remove' )
```

 CATEGORICAL DATA ENCODING

```python
def DataEncoding(df):
    encoded_df = pd.get_dummies(df)
    return encoded_df
```

```python
e_df = DataEncoding(df)
```
